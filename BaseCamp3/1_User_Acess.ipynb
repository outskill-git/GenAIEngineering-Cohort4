#### User Data Access ####  
**APIs** provided by the User_Management Module is used to perform actions  
Its kept modular way - abstracted details (DB details, complete data, internal mechanism not visible)

```python
from User_Management import read_user, add_user, update_user
```

**Read**  
Read existing data based on User Id

```python
# Read User details with User Id
user = read_user("U_0011")
print(user)
```

**Update**  
Update details of existing user

```python
msg = update_user("U_0002",  city="New Delhi", email="example.new@gmail.com")
print(msg)
```

```python
print("After update :", read_user("U_0002"))
```

**Add**  
Add new user record into the DB  
Read the details after added to check

```python
new = {
    "name": "Asha Patel",
    "age": 29,
    "city": "Ahmedabad",
    "email": "asha.patel@example.com",
    "phone_number": "+91-98765-43210"
}
new_id = add_user(new)
print("Added new user id:", new_id)
print("New user :", read_user(new_id))
```
